# 01 — Dataset Exploration

Explore the RadioML 2018.01A dataset structure, distributions, and sample waveforms.

## Prerequisites

Place the dataset at `data/raw/GOLD_XYZ_OSC.0001_1024.hdf5` or update the path below.

```python
import sys
sys.path.insert(0, '../src')

import numpy as np
import h5py
import matplotlib.pyplot as plt

DATASET_PATH = '../data/raw/GOLD_XYZ_OSC.0001_1024.hdf5'
```

```python
# Inspect HDF5 structure
with h5py.File(DATASET_PATH, 'r') as f:
    for key in f.keys():
        print(f'{key}: shape={f[key].shape}, dtype={f[key].dtype}')
```

```python
# Load a subset for exploration
from rf_sentinel.config.loader import load_config
from rf_sentinel.data.radioml_loader import load_radioml

cfg = load_config('../configs/cnn1d_quick_test.yaml')
dataset = load_radioml(cfg)
print(dataset)
```

```python
# Plot sample I/Q waveform
idx = 0
sample = dataset.X[idx]
I, Q = sample[:, 0], sample[:, 1]

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
axes[0].plot(I, label='I')
axes[0].plot(Q, label='Q')
axes[0].set_title(f'I/Q — {dataset.class_names[dataset.labels[idx]]} (SNR={dataset.snr[idx]}dB)')
axes[0].legend()

amp = np.sqrt(I**2 + Q**2)
axes[1].plot(amp, color='green')
axes[1].set_title('Amplitude')

axes[2].scatter(I, Q, s=1, alpha=0.5)
axes[2].set_title('Constellation')
axes[2].set_aspect('equal')
plt.tight_layout()
plt.show()
```
